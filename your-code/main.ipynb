## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
mean = 10 # minutes
num_of_events = 15 # minutes

less_than_15 = stats.poisson.cdf(num_of_events, mean)
print(f"The probability of a customer spending less than {num_of_events} is {less_than_15:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
more_than_15 = 1- less_than_15
print(f"The probability of a customer spending more than {num_of_events} is {more_than_15:.4f}")
```
